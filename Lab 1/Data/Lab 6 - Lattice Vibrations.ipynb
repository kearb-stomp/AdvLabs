```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#data importing function
def load_data(filepath):
    #open the file and determine where the data starts using the #DATA string in the file
    with open(filepath, 'r') as f:
        lines = f.readlines()

    data_start = None

    #To check the line reading function
    #with open(filepath, 'r') as f:
     #   lines = f.readlines()
    
    for idx, line in enumerate(lines):
        if line.strip().startswith('#DATA'):  # == '#DATA':
            data_start = idx + 1  #This sets the data start line to the one following #DATA

            break

    if data_start is None:
        raise ValueError("No #DATA section found in file.")

    #Replace commawith period for decimal separators
    lines = [line.replace(',', '.') for line in lines[data_start:]]

    #This loads the data from the file
    data = np.loadtxt(lines[data_start:], delimiter='\t')

    #Split the data into two separate arrays for ease of use
    wavenumber = data[:,0]
    intensity = data[:,1]

    return wavenumber, intensity
```

```python
#load the transmissivity data for water vapor calibration
vapor_twavenumber, vapor_tintensity = load_data(r'Data/Transmittance/Water-vapor.ASC')

#quick plot to check the data importation
plt.plot(vapor_twavenumber, vapor_tintensity)
plt.xlabel('Wavenumber (cm$^{-1}$)')
plt.ylabel('Intensity (arb. units)')
plt.title('Calibration Spectrum - Water Vapor')
plt.gca().invert_xaxis()
plt.show()
```
